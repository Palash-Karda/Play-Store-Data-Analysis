```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Step 1: Prepare the dataset
# Features: IMDb rating and Duration
X = np.array([[8.0, 160], [6.2, 170], [7.29, 168], [8.2, 155]])

# Labels: Genre (0 = Action, 1 = Comedy)
y = np.array([0, 0, 1, 1])  # 0 = Action, 1 = Comedy

# Step 2: Split the data into training and test sets
# We'll use 75% of the data for training and 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Step 3: Create and train the k-NN classifier with k=3
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)

# Step 4: Test the model using the test data and calculate accuracy
y_pred = knn.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy on the test set: {accuracy * 100:.2f}%")

# Step 5: Predict the genre for the new movie (Barbie)
barbie_movie = np.array([[7.4, 114]])  # IMDb rating = 7.4, Duration = 114 minutes
predicted_genre = knn.predict(barbie_movie)

# Step 6: Output the predicted genre
if predicted_genre[0] == 0:
    print("The predicted genre for Barbie is: Action")
else:
    print("The predicted genre for Barbie is: Comedy")
```
